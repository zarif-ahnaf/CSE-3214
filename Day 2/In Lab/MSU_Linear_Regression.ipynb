```python
X=data['Height(FT)'].values
Y=data['Weight(KG)'].values
```

```python
meanX=X.mean()
meanY=Y.mean()
```

```python
print(meanX)
print(meanY)
```

```python
n=len(X)
```

```python
print(n)
```

```python
import numpy as np
```

```python
m=(((X-meanX)*(Y-meanY)).sum())/((((X-meanX)**2)).sum())
print(m)
```

```python
c=meanY-(m*meanX)
```

```python
print(c)
```

```python
new_x=float(input("Enter Height:"))
#take unkonwn input
y=(m*new_x)+c
print("Possible Weight:",y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(4,8,100) #(4-8)-->range
y=(m*x)+c
plt.plot(x,y,color="g",label="Regression Line")
plt.scatter(X,Y, color="r",label="Scatter Plot")
plt.title("Regression")
plt.xlabel("Height (FT)")
plt.ylabel("Weight (KG)")
plt.legend()
plt.show()
```
